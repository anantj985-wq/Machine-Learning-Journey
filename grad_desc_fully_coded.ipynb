```python
import numpy as np
```

```python
x=np.random.randn(10,1)
print(x)
```

```python
y=4 * x + np.random.rand()
#parameter
w=0
b=0
```

```python
#hyperparameter
lr=0.01
```

```python
# 1. Update your function cell
def grad_desc(x, y, w, b, lr):
    dldw = 0.0
    dldb = 0.0
    
   
    N = x.shape[0] 
    
    for xi, yi in zip(x, y):
        dldw += -2 * xi * (yi - (w * xi + b))
        dldb += -2 * (yi - (w * xi + b))
        
    w = w - lr * (1 / N) * dldw
    b = b - lr * (1 / N) * dldb
    
    # Using float() converts any leftover 1-element numpy arrays into normal numbers
    return float(w), float(b)
```

```python
# 2. Run your iteration cell
for epoch in range(400):
    w, b = grad_desc(x, y, w, b, lr)
    
    yhat = w * x + b
    # Also change .shape(0) to .shape[0] here if you use it to calculate loss average
    loss = np.divide(np.sum((y - yhat) ** 2), x.shape[0])
    
    if epoch % 50 == 0:
        print(f'Epoch {epoch}: loss is {loss:.4f}, parameters w: {w:.4f}, b: {b:.4f}')
```
