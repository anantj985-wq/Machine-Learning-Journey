```python
df = df.drop(columns=['User ID'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])
```

```python
X = df[['Age','EstimatedSalary']]

y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
import numpy as np

def euclidean_distance(point1, point2):

    distance = np.sqrt(
        np.sum((point1 - point2)**2)
    )

    return distance
```

```python
test_point = X_test_scaled[0]
```

```python
distances = []

for train_point in X_train_scaled:

    d = euclidean_distance(
        test_point,
        train_point
    )

    distances.append(d)
```

```python
distances = np.array(distances)

nearest_indices = np.argsort(distances)
```

```python
k = 5
k_indices = nearest_indices[:k]
```

```python
k_labels = y_train.iloc[k_indices]
```

```python
prediction = np.bincount(
    k_labels
).argmax()
```

```python
def predict_one(test_point,k):

    distances = []

    for train_point in X_train_scaled:

        d = euclidean_distance(
            test_point,
            train_point
        )

        distances.append(d)

    nearest = np.argsort(distances)[:k]

    labels = y_train.iloc[nearest]

    prediction = np.bincount(labels).argmax()

    return prediction
```

```python
predictions = []

for point in X_test_scaled:

    pred = predict_one(
        point,
        k=5
    )

    predictions.append(pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(
    y_test,
    predictions
)
```
